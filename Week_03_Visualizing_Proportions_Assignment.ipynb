```python
cols = data_mushrooms.select_dtypes(["object"]).columns
data_mushrooms[cols] = data_mushrooms[cols].astype('category')
```

```python
edibleclass=data_mushrooms.groupby(['class']).count()
edibleclass
```

```python
data_mushrooms.head(10)
```

# Pie

```python
labels=['Edible','Poisonous']
plt.pie(edibleclass['population'],labels=labels,autopct='%.1f %%')
plt.title('Edible?')
plt.show()
```

# Donuts

```python
habitat=data_mushrooms.groupby(['habitat']).count()
```

```python
habitat.head()
```

```python
labels=['Grasses','Leaves','Meadows','Paths','Urban','Waste','Wood']

plt.pie(habitat['class'], labels=labels,
        autopct='%1.1f%%', pctdistance=0.85) 
center_circle = plt.Circle((0, 0), 0.50, fc='white')
fig = plt.gcf()

fig.gca().add_artist(center_circle)
  
plt.title('Mushroom Habitats')
  
plt.show()
```
